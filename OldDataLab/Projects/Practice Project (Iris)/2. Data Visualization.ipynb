##3. Visualize dataset features - multivariate plots

###3.1 Correlation Matrix Plot

A correlation matrix calculates the correlation between each pair of variables. The correlation matrix plot is useful in finding highly correlated pairs of variables.

Compute the correlation matrix using the `corr()` dataframe method.

```
correlations = import_df.corr()
```

Four steps tp plot the correlation matrix:
1. Set up figure with colorbar
1. Set up axes
1. Display the correlation matrix
1. Set up ticks and show label at every tick

For illustration purpose, the four steps will be split into two parts as seen below.

__Step 1 - 3__
In the cell below, some methods and functions in the `matplotlib.pyplot` are used:
 
- `figure()` is used to create a figure which is the overall window or page that everything is drawn on.
- `add_subplot()` is used to add 'axes' to the figure. (Parameter `111` is equal to `1,1,1`, which is to give three arguments to `add_subplot()`. The three arguments designate the number of rows (1), the number of columns (1) and the plot number (1). So `111` means to make one subplot.)
- `matshow()` is used to display the correlation matrix
- `colorbar()` is used to add a colorbar to the plot

```
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1)
fig.colorbar(cax)

display(plt.show())
```

__Step 4__
To set up ticks and display label at every tick, we need to call several functions:

- `set_xticks()` is used to set the x ticks. Here we use `np.arange(0,4,1)` to generate a list `[0,1,2,3]` as the locations for x-axis ticks.
- `set_yticks()` is used to set the y ticks.
- `set_xticklabels()` is used to set the x-tick labels. Here we use `list(import_df)` as a list of string labels.
- `set_yticklabels()` is used to set the y-tick labels.

```
ticks = np.arange(0,4,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)

ax.set_xticklabels(list(import_df))
ax.set_yticklabels(list(import_df))

display(plt.show())
```
