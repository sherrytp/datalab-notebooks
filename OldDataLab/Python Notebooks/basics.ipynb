```
iris_df.plot.hist(x='SepalLength', bins=20, range=(0,10), normed=False)
plt.xlabel('Sepal Length (cm)')
display(plt.show())
```

```
iris_df.plot.scatter(x='SepalLength', y='SepalWidth')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
display(plt.show())
```

```
iris_df.plot(x='SepalLength', y='SepalWidth', kind='hexbin')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
display(plt.show())
```
