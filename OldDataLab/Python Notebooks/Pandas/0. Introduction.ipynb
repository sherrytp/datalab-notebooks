# Introduction

The `sklearn.preprocessing` package provides several common utility functions and transformer classes to change raw feature vectors into a representation that is more suitable for the downstream estimators. The `sklearn.preprocessing` module includes scaling, centering, normalization, binarization and imputation methods.
