```
import_df = read_dataframe()
import_df\
  .assign(drive_wheels_alt = lambda df: df.drive_wheels.str.replace('rwd','rear')\
                                                       .str.replace('fwd','front')\
                                                       .str.replace('4wd','all')
         )\
  .loc[:,['drive_wheels_alt','drive_wheels']]\
  .head()
```

```
import_df.body_style.str.match('convertible')
```

```
import_df = read_dataframe()
import_df\
  .assign(is_convertible = lambda df: df.body_style.str.match('convertible').astype('int64'))\
  .loc[:,['is_convertible','body_style']]\
  .head()
```

```
import_df = read_dataframe()
import_df\
  .assign(engine_size_l3 = import_df.engine_size*2.54**3/10**3)\
  .rename(columns={'engine_size':'engine_size_i3'})\
  .loc[:,['engine_size_i3','engine_size_l3']]\
  .head()
```
