## 1. Setup

```
wdi_part1of2_filepath = 'https://github.com/datalab-datasets/world-development-indicators/blob/master/WDIData_part1of2.csv?raw=true'
wdi_part2of2_filepath = 'https://github.com/datalab-datasets/world-development-indicators/blob/master/WDIData_part2of2.csv?raw=true'
```

### `wdi` using `read_csv` from pandas (Python)

- http://pandas.pydata.org/pandas-docs/version/0.19/generated/pandas.read_csv.html

```
import pandas as pd
pd.__version__
```

```
import pandas as pd
wdi_part1of2_pdf = pd.read_csv(wdi_part1of2_filepath)
wdi_part2of2_pdf = pd.read_csv(wdi_part2of2_filepath)
```

```
wdi_data_pdf = \
pd.concat([wdi_part1of2_pdf,
           wdi_part2of2_pdf],
          axis=0
          ) \
  .rename(columns={'Country Code':'country_code', 
                   'Indicator Code':'indicator_code'}) \
  .drop(axis=1, 
        labels=['Country Name',
                'Indicator Name',
                'Unnamed: 61']) \
  .set_index(['country_code',
              'indicator_code'])
```
