### Check Stationary 
Plotting Rolling Statistics: The function will plot the moving mean or moving Standard Deviation. 
Dickey-fuller Test: Statistical hypothesis tests for checking stationarity 
* $H_0$ is non-stationary, to reject: t-stats less than critical value 

```python
def test_stationarity(timeseries):
    
    #Determing rolling statistics
    rolmean = timeseries.rolling(12).mean()
    rolstd = timeseries.rolling(12).std()
    
    #Plot rolling statistics:
    plt.plot(timeseries, color='blue',label='Original')
    plt.plot(rolmean, color='red', label='Rolling Mean')
    plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show()
    
    #Perform Dickey-Fuller test:
    print('Results of Dickey-Fuller Test:')
    dftest = sm.tsa.stattools.adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in list(dftest[4].items()):
        dfoutput['Critical Value (%s)'%key] = value
    print(dfoutput) 
    
test_stationarity(CN_wealth[109:])
```

Not stationary: 
1. mean is not constant even though st. dev is small 
2. t-stats is so much larger than critical value 

```python
# Decomposition 
#from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = sm.tsa.seasonal_decompose(CN_wealth[109:])

trend = decomposition.trend
residual = decomposition.resid 
plt.subplot(211)
plt.plot(trend, label = "Trend")
plt.legend(loc = "best")
plt.subplot(212)
plt.plot(residual, label = "Residuals")
plt.legend(loc = "best")
plt.tight_layout()
```

```python
# remove trend and check stationary again 
test_stationarity(residual)
```

```python
mod = sm.tsa.statespace.SARIMAX(CN_wealth[109:], 
                                order = (1, 1, 1),
                                seasonal_order = (1, 1, 0, 12),
                                enforce_stationarity = False,
                                enforce_invertibility = False)
results = mod.fit()
print(results.summary().tables[1])
results.plot_diagnostics(figsize = (16, 8))
plt.show()
```

Position with Endogenous and directed technological change to describe the innovation possibilities frontier (improving technology and also creating new technologies). 

**Assume**: 
* There is a fixed (inelastic) supply of $\mathbb{S}$ scientists in the economy. 
* Advances in automation and creation of new technologies are given, where we need to measure the easement or difficulty of the technological change 

The key object is to solve the value function, determining the net present discounted value of new automation $V_I(t)$ and labor-intensive innovations $V_N(t)$. 
$V_I(t)$ is the value of a new technology automating the task at $i = I(t) + \epsilon$ (the highest-indexed task that hasn't been automated), $V_N(t)$ is the value of a new technology creating a more complex task at $i = N(t) + \epsilon$. 

The equilibium of the value function with endogenous technology is given by paths $\{K(t), N(t), I(t)\}^{\infty}_{t=0}$ for capital and technology (starting from an initial values $K(0), N(0), I(0)$), paths $\{r(t), W(t), W^S(t)\}^{\infty}_{t=0}$ for factor prices, paths $\{V_N(t), V_I(t)\}^{\infty|_{t=0}$ for the value functions, and paths $\{S_N(t), S_I(t)\}^^{\infty}_{t=0}$ for the allocation of scientists when the market is clear and all firms, including the prospective technology monopolists, maximize profits, the representatitive household maximizes its utility. 

Flow profits from automation: 
$\pi_I(t,i) = (1-\mu)(\frac{\eta}{1-\eta})^\zeta\psi^{1-\zeta} Y(t)c^u(r(t))^{\zeta - \sigma}$

Flow profits from frims producing task $i$ using labor: 
$\pi_N(t,i) = (1-\mu)(\frac{\eta}{1-\eta})^\zeta\psi^{1-\zeta} Y(t)c^u(\frac{W(t)}{\gamma(i)})^{\zeta - \sigma}$ 

The optimization will reach by the net present discounted value of the profit stream, discounted using the path of future interest rates, that the existing patent-hold would obtain 

$(1-\mu)(\frac{\eta}{1-\eta})^{\zeta}\psi^{1-\zeta} \int^{\infty}_t e^{-\int^\tau_0(r(s)-\sigma)ds}Y(\tau)c^u(\frac{W(\tau)}{\gamma(I)})^{\zeta - \sigma} d\tau$. 

So the values of firms that innovate with automation and creation of new tasks will be: 

$V_I(t) = (1-\mu)(\frac{\eta}{1-\eta})^{\zeta}\psi^{1-\zeta} \int^{\infty}_t e^{-\int^\tau_t(r(s)-\sigma)ds}Y(\tau)(c^u(r(\tau))^{\zeta - \sigma} - c^u(\frac{W(\tau)}{\gamma(I(t))})^{\zeta - \sigma})d\tau$.
